```python
import time

messages = []

def write_message():
    msg = input("\nWrite your message for the future: ")
    delay = int(input("Set a time delay (in seconds) before it unlocks: "))
    messages.append((msg, time.time() + delay))
    print("🕰️ Message saved! Check back later to see it.")

def view_messages():
    current_time = time.time()
    unlocked = [msg for msg, unlock_time in messages if current_time >= unlock_time]

    if unlocked:
        print("\n📜 Unlocked Messages:")
        for msg in unlocked:
            print(f"- {msg}")
        # Remove unlocked messages
        messages[:] = [(msg, t) for msg, t in messages if t > current_time]
    else:
        print("\n🔒 No messages unlocked yet! Be patient.")

while True:
    print("\n⏳ Time Capsule Messenger")
    print("1. Write a Message\n2. View Messages\n3. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        write_message()
    elif choice == "2":
        view_messages()
    elif choice == "3":
        print("See you in the future! 🚀")
        break
    else:
        print("Invalid choice! Try again.")
```
